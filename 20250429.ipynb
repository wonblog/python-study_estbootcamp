```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 2, 301)
y1 = 3**x
y2 = 3**(x-1)+2

graph1, = plt.plot(x,y1)
graph2, = plt.plot(x,y2)
print(type(graph1))
plt.legend(handles=(graph1,graph2), labels=(r'$y=3^x$', r'$y=3^{x-1}+2$'))
plt.show()
```

```python
x1=np.linspace(0.1,4,400)
y1=np.log2(x1)
x2=np.linspace(-0.9,4,500)
y2=np.log2(x2+1)-1
graph1, = plt.plot(x1,y1)
graph2, = plt.plot(x2,y2)
print(type(graph1))
plt.legend(handles=(graph1,graph2), labels=(r'$y=log_2(x)$', r'$y=log_2(x+1)-1$'))
plt.show()
```
